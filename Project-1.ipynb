```python
######################################## CODE #############################################

#Importing important lbraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Reading the data from csv file and storing it in a variable
customers=pd.read_csv('../input/ecommerce-customers/Ecommerce Customers.csv')

#Training and Testing Data
y=customers['Yearly Amount Spent']
x=customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)

#Training 
lm=LinearRegression()
lm.fit(X_train,y_train)

#Declaring a variable and storing in it the coefficients from the training data
coef=lm.coef_
#Predicting test data
predictions=lm.predict(X_test)




############################# OUTPUT CODE ####################################

#The coefficients with respect to their corresponding coloumns in the data
cdf=pd.DataFrame(lm.coef_,x.columns,columns=['coeff'])
print('\nThe coefficients with respect to their corresponding coloumns : \n',cdf)

#scatter plot of the real test values versus the predicted values.
print('\nScatter plot of the real test values versus the predicted values is giben at the bottom of the output screen.\n')
plt.scatter(y_test,predictions)
plt.xlabel('Y Test(True values)')
plt.ylabel('Predicted values')




########################### Model Evaluation ####################################

#evaluating the model performance by calculating the residual sum of squares.
from sklearn import metrics
print('The residual sum of squares are giben below :\n')
print('MAE',metrics.mean_absolute_error(y_test,predictions))
print('MSE',metrics.mean_squared_error(y_test,predictions))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
